```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bf_data=pd.read_csv('BlackFriday.csv')
```

```python
bf_data.head()
```

```python
bf_data.describe()
```

```python
bf_data.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis=1,inplace=True)
```

```python
bf_data.head()
```

```python
bf_data.City_Category=bf_data.City_Category.fillna('A')
```

```python
bf_data.head()
```

```python
bf_data.City_Category=bf_data.City_Category.map({"A":"Metro Cities","B":"Small Towns","C":"Villages"})
```

```python
bf_data.head()
```

```python
bf_data.rename(columns={'Product_Category_1':'Baseball Caps', 'Product_Category_2':'Wine Tumblers','Product_Category_3':'Pet Raincoats'},inplace=True)
```

```python
bf_data.head()
```

```python
bf_data.Marital_Status=bf_data.Marital_Status.map({1:'Married',0:'Unmarried'})
```

```python
bf_data.head()
```

```python
print(pd.crosstab(bf_data.Gender,bf_data.City_Category))
```

```python
ax=sns.countplot(x='Gender',hue='City_Category',data=bf_data,palette='Set1')
ax.set(title='Gender vs City Category')
plt.show()
```

```python
pd.crosstab(bf_data.Gender,bf_data['Baseball Caps'])
```

```python
sns.countplot(x='Gender',hue='Baseball Caps',data=bf_data)\
.set(title='Gender vs Baseball Caps Category')
plt.show()
```

```python
sns.countplot(x='Gender',hue='Wine Tumblers',data=bf_data)\
.set(title='Gender vs Wine Tumblers Category')
plt.show()
```
