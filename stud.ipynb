```python
import numpy as np

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

student_df = pd.read_csv('StudentsPerformance.csv')
student_df.drop(['lunch','test preparation course',], axis=1, inplace=False)
print(student_df.head(5))

student_df ['race/ethnicity'] = student_df ['race/ethnicity'].map({
    'group A': 'Asian Students',
    'group B': 'African Students',
    'group C':'Afro-Asian Students',
    'group D':'American Students',
    'group E':'European Students'
    
})
student_df["parental level of education "] = student_df["parental level of education"].fillna("NE")
print(student_df.head(5))

intervals=(0,40,50,75,100)
grades=['failed','second class','first class','distinction']
student_df['grades']=pd.cut(student_df.mean(axis=1),intervals,labels=grades)
ax=sb.countplot(x='gender',data=student_df,hue='grades',palette='Set1')\
.set(xlabel=' gender',ylabel='grades',Title='performance')
plt.show()
ax=sb.countplot(x='test preparation course',data=student_df,hue='gender',palette='Set2')\
.set(xlabel='course',ylabel='pass/fail',Title='graph1')
plt.show()
ax=sb.countplot(x='race/ethnicity',data=student_df,hue='gender',palette='Set3')\
.set(xlabel='race',ylabel='students',Title='graph2')
plt.show()

```
